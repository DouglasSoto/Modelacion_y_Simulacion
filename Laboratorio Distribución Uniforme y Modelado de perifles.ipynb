```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "C:/Users/PC/Downloads/Python"
file = "datosincompletos2020.csv"
data = pd.read_csv(path + "/" + file)
```

```python
data
```

```python
cantidadatos = 25
```

```python
edad = np.random.uniform(20,40,cantidadatos) + 4*np.random.randn(cantidadatos)
```

```python
experiencia = np.random.uniform(2,20,cantidadatos)
```

```python
estudio = np.random.uniform(2,5,cantidadatos)
```

```python
Tabla = pd.DataFrame(
    {
        'edad':edad,
        'años de experiencia':experiencia,
        'Años de estudio':estudio
    }
)
```

```python
Tabla
```

```python
tablasunidas = pd.concat([data, Tabla], axis=1)
tablasunidas
```

```python
salarios = []

ingresosList = tablasunidas["ingresos"].values.tolist()
maxSalario = max(ingresosList)
minSalario = min(ingresosList)
salarios.append(max(ingresosList))
salarios.append(min(ingresosList))
%matplotlib inline
plt.hist(salarios)
```

```python
tablasunidas.describe()
```

```python
#interpretación

#La desviación estándar de los datos de los ingresos es en promedio de 2400.
#Los salarios pueden variar dependiendo de la profesión y el puesto de trabajo.
#Los salarios pueden variar dependiendo de los años de experiencia y los años de estudio.
```

```python
sns.set_theme(color_codes=True)
sns.regplot(x="edad", y="ingresos", data=tablasunidas);
```

```python
# interpretacion

#En la grafica se muestra que no necesariamente la edad es un factor importante para obtener más ingresos
```

```python
sns.set_theme(color_codes=True)
sns.regplot(x="años de experiencia", y="ingresos", data=tablasunidas);
```

```python
# interpretacion

#En la grafica se muestra que los años de experiencia tampoco son un factor importante para obtener más ingresos
```
